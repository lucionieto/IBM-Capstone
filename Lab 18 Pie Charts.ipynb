### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Drop NaNs in the column
df_db = df.dropna(subset=['DatabaseWantToWorkWith'])

# Split multiple databases per respondent
df_db['DatabaseWantToWorkWith'] = df_db['DatabaseWantToWorkWith'].str.split(';')
df_exploded = df_db.explode('DatabaseWantToWorkWith')

# Count frequency of each database
db_counts = df_exploded['DatabaseWantToWorkWith'].value_counts()

# Get top 5 most desired databases
top_databases = db_counts.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top 5 Most-Desired Databases')
plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Drop NaNs in DevType
df_dev = df.dropna(subset=['DevType'])

# Split multiple roles per respondent
df_dev['DevType'] = df_dev['DevType'].str.split(';')
df_exploded = df_dev.explode('DevType')

# Count frequency of each role
role_counts = df_exploded['DevType'].value_counts()

# Get top 5 roles
top_roles = role_counts.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_roles, labels=top_roles.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Developer Roles')
plt.tight_layout()
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


```python
import pandas as pd
import matplotlib.pyplot as plt

# Drop rows with missing OpSysProfessionalUse
df_os = df.dropna(subset=['OpSysProfessional use'])

# Split multiple OS entries into lists
df_os['OpSysProfessional use'] = df_os['OpSysProfessional use'].str.split(';')

# Explode into separate rows
df_exploded = df_os.explode('OpSysProfessional use')

# Count each OS
os_counts = df_exploded['OpSysProfessional use'].value_counts()

# Select top 5 OS types (or however many you want to show)
top_os = os_counts.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set2.colors)
plt.title('Top Operating Systems Used Professionally')
plt.tight_layout()
plt.show()
```

##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Drop missing values
df_lang = df.dropna(subset=['LanguageHaveWorkedWith'])

# Split multiple language entries
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')

# Explode into separate rows
df_exploded = df_lang.explode('LanguageHaveWorkedWith')

# Count top 5 languages
top_langs = df_exploded['LanguageHaveWorkedWith'].value_counts().nlargest(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_langs, labels=top_langs.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Programming Languages Used by Respondents')
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```
