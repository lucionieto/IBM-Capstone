### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
print (df['Country'].unique())
print(df['EdLevel'].unique())
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
country_mapping = {
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Russian Federation': 'Russia',
    'Iran, Islamic Republic of...': 'Iran',
    'Republic of Korea': 'South Korea',
    "Democratic People's Republic of Korea": 'North Korea',
    'Venezuela, Bolivarian Republic of...': 'Venezuela',
    'Libyan Arab Jamahiriya': 'Libya',
    'Micronesia, Federated States of...': 'Micronesia',
    'Congo, Republic of the...': 'Congo',
    'Hong Kong (S.A.R.)': 'Hong Kong',
    "Lao People's Democratic Republic": 'Laos',
    'Republic of North Macedonia': 'North Macedonia',
    'Republic of Moldova': 'Moldova',
    'United Republic of Tanzania': 'Tanzania'
}

df['Country'] = df['Country'].replace(country_mapping)

education_mapping = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor’s degree',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master’s degree',
    'Some college/university study without earning a degree': 'Some college',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High school',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Doctorate or professional degree',
    'Associate degree (A.A., A.S., etc.)': 'Associate degree',
    'Primary/elementary school': 'Primary school',
    'Something else': 'Other'
}

df['EdLevel'] = df['EdLevel'].replace(education_mapping)

```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
##  Identify columns with the highest number of missing values.

for col in df.columns:
    missing_df = df[col].isnull().sum()

missing_counts = df.isnull().sum()
missing_sorted = missing_counts[missing_counts > 0].sort_values(ascending=False)

print("Columns with the Most Missing Values:\n")
print(missing_sorted)
 
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here

frequent = (df['ConvertedCompYearly'].median())
print(frequent)

df['ConvertedCompYearly'].fillna(frequent, inplace = True )

print(df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here


frequent = (df['RemoteWork'].mode()[0])
print(frequent)

df['RemoteWork'].fillna(frequent, inplace = True )

print(df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df['ConvertedCompYearly_MinMax'] = df['ConvertedCompYearly']/df['ConvertedCompYearly'].max()
print(df['ConvertedCompYearly_MinMax'])
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np
df['log_Comp'] = np.log1p(df['ConvertedCompYearly'])  # log1p handles zero values safely
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
## Write your code here
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

def map_experience(years):
    if pd.isnull(years):
        return 'Unknown'
    elif years <= 1:
        return 'Beginner (≤1 yr)'
    elif years <= 3:
        return 'Novice (2–3 yrs)'
    elif years <= 5:
        return 'Intermediate (4–5 yrs)'
    elif years <= 8:
        return 'Skilled (6–8 yrs)'
    elif years <= 12:
        return 'Advanced (9–12 yrs)'
    elif years <= 20:
        return 'Expert (13–20 yrs)'
    else:
        return 'Veteran (21+ yrs)'

# Step 3: Apply mapping to create ExperienceLevel
df['ExperienceLevel'] = df['YearsCodePro'].apply(map_experience)

# Step 4: Preview the result
print(df[['YearsCodePro', 'ExperienceLevel']].head())
```
